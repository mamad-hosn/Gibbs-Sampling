```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu, sigma = 9.4, 7.8 ** 0.5
normal_sample = np.random.normal(mu, sigma, 1000)
var = np.std(normal_sample) ** 2
plt.hist(normal_sample, bins = 30)
plt.show()
print(var)
```

```python
x_axis = np.arange(9.4 - 15, 9.4 + 15, 0.01)
first_data = np.array([3, 9, 6, 6, 16, 6, 8, 10, 9, 12, 10, 9, 13, 6, 10, 7,
             10, 4, 11, 11, 10, 4, 10, 10, 12, 10, 13, 12, 10, 12, 
             11, 10, 11, 10, 10, 6, 16, 11, 11, 10, 10, 11, 5, 11, 
             5, 9, 10, 10, 10, 10, 3, 7, 12, 11, 6, 8, 10, 12])

plt.hist(first_data, bins = 14, density = True, label = 'first data Dist')
plt.plot(x_axis, norm.pdf(x_axis, mu, sigma), color = 'Black', label = 'normal PDF')
plt.legend(loc='upper right')
plt.show()
```
